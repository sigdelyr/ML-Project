```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import json
import urllib.request

QUERY_URL = "https://www.alphavantage.co/query?function={REQUEST_TYPE}&apikey={KEY}&symbol={SYMBOL}&outputsize=compact"
API_KEY = "GDFMG7QXIZS4308T"

def _request(symbol, req_type):
    with urllib.request.urlopen(QUERY_URL.format(REQUEST_TYPE=req_type, KEY=API_KEY, SYMBOL=symbol)) as req:
        data = req.read().decode("UTF-8")
    return data

def get_daily_data(symbol):
    return json.loads(_request(symbol, 'TIME_SERIES_DAILY'))
```

```python
json_data = get_daily_data('AAPL')
```

```python
json_data.keys()
```

```python
# convert json file into dataframe(apple)
apple_df = pd.DataFrame(json_data['Time Series (Daily)']).T
apple_df.head()
```

```python
apple_df.rename_axis("date", axis='index', inplace=True)
```

```python
apple_df = apple_df.rename(columns={'1. open':'open', '2. high': 'high', '3. low': 'low', '4. close': 'close', '5. volume': 'volume'})
apple_df.head()
```

```python
# load CHK(cheasapeak energy) stock data into pandas dataframe
json_df = get_daily_data('CHK')
```

```python
json_df.keys()
```

```python
chk_df = pd.DataFrame(json_df['Time Series (Daily)']).T
chk_df.head()
```

```python
chk_df.rename_axis("date", axis='index', inplace=True)
```

```python
chk_df = chk_df.rename(columns={'1. open':'open', '2. high': 'high', '3. low': 'low', '4. close': 'close', '5. volume': 'volume'})
chk_df.head()
```

```python
# merge two dataframe on date column
stock_df = (pd.merge(apple_df, chk_df, on = "date", how='left')
        .fillna(0))
stock_df.head()
```

```python
stock_df = stock_df.rename(columns={'open_x':'Aopen', 'high_x': 'Ahigh', 'low_x': 'Alow', 'close_x': 'Aclose',
                                    'volume_x': 'Avolume', 'open_y': 'Copen', 'high_y': 'Chigh', 'low_y': 'Clow', 'close_y': 'Cclose', 'volume_y': 'Cvolume'})
stock_df.head()
```
