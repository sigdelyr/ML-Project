```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime
import time
```

```python
df = pd.read_csv('resources/apple_stock.csv', sep = ",")
```

```python
# Convert Date column to datetime
df.Date=pd.to_datetime(df.Date).dt.strftime('%Y-%m-%d')
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df.Date,format='%Y-%m-%d')
df['Year'] = df['Date'].dt.year
df['Quarter'] = df['Date'].dt.quarter
df.head()
```

```python
apple_df = df.rename(columns={'Adj Close':'Adj_Close'})
apple_df.tail()
```

```python
apple_df.columns
```

```python
apple_df.to_csv('../ML-Project/apple.csv',index=True)
```

```python
# Merge csv's and Read the csv file into a pandas DataFrame
inc = pd.read_csv('is.csv')
apple = pd.read_csv('apple.csv')

merged = inc.merge(apple, on= ['Quarter', 'Year'], how = 'outer')
merged.head()
```

```python
merged = merged.rename(columns={"Date_x": "Quarterly Dates",
                               "Date_y": "Daily Dates"})
merged.columns
```

```python
merged.drop(['Unnamed: 0'], axis=1, inplace = True)
merged.head()
```

```python
merged.to_csv('../ML-Project/final.csv',index=True)
```
