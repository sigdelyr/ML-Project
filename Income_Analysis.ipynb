```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Merge csv's and Read the csv file into a pandas DataFrame
inc = pd.read_csv('is.csv')
apple = pd.read_csv('apple.csv')

merged = inc.merge(apple, on= ['Quarter', 'Year'], how = 'outer')
```

```python
merged.head()
```

```python
merged = merged.rename(columns={"Date_x": "Quarterly Dates",
                               "Date_y": "Daily Dates"})
merged.columns
```

```python
merged.drop(['Unnamed: 0'], axis=1, inplace = True)
merged.head()
```

```python
merged.to_csv('apple.csv',index=True)
```

```python
# Assign the data to X and y
# Note: Sklearn requires a two-dimensional array of values
# so we use reshape to create this
X = inc["Net Income"].values.reshape(-1, 1)
y = inc["Profit"].values.reshape(-1, 1)

print("Shape: ", X.shape, y.shape)
```

```python
 # Plot the data to see if a linear trend exists

plt.scatter(X, y)
plt.xlabel("Net Income")
plt.ylabel("Profit")
```

```python
# Use sklearn's `train_test_split` to split the data into training and testing

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create the model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# Fit the model to the training data. 
model.fit(X_train, y_train)
```

```python
# Calculate the mean_squared_error and the r-squared value
# for the testing data

from sklearn.metrics import mean_squared_error, r2_score

# Use our model to make predictions
predicted = model.predict(X_test)

# Score the predictions with mse and r2
mse = mean_squared_error(y_test, predicted)
r2 = r2_score(y_test, predicted)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2 ): {r2}")
```
